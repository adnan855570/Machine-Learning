```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Titanic.csv')
df.head()
```

# **1. Categorical Data**

## a. Countplot

```python
sns.countplot(['survived'])
```

```python
df['survived'].value_counts().plot(kind='bar')
```

```python
df['survived'].value_counts().plot(kind='pie')
```
